```python
from ultralytics import YOLO
import matplotlib.pyplot as plt
import numpy as np
import cv2
# Load a model
model = YOLO('yolov8n.pt')  # pretrained YOLOv8n model

# Run batched inference on a list of images
image_paths = ['image1.png', 'image1.jpeg', 'image2.jpg']
results = model(image_paths, stream=True)  # return a generator of Results objects

# Process results generator
for result, image_path in zip(results, image_paths):
    # Extract relevant information from the Results object
    boxes = result.xyxy[0].cpu().numpy()
    confidences = result.conf[0].cpu().numpy()
    class_ids = result.xyxy[0][:, -1].cpu().numpy()

    # Load the image
    img = plt.imread(image_path)

    # Display the image with bounding boxes
    plt.imshow(img)

    # Draw bounding boxes
    for box, confidence, class_id in zip(boxes, confidences, class_ids):
        x1, y1, x2, y2 = box[:4]
        plt.rectangle((x1, y1), x2 - x1, y2 - y1, linewidth=2, edgecolor='r', facecolor='none')
        plt.text(x1, y1, f'{model.names[int(class_id)]}: {confidence:.2f}', bbox=dict(facecolor='red', alpha=0.5))

    plt.title(f"Result for {image_path}")
    plt.show()


output_path = f"annotated_{image_path}"
cv2.imwrite(output_path, img)

    # Display the annotated image (optional)
cv2.imshow('Annotated Image', img)
cv2.waitKey(0)
cv2.destroyAllWindows()

print(f"Annotated image saved at: {output_path}")
```

```python
import cv2
from ultralytics import YOLO
import pafy
import numpy as np
import math
import threading
import queue

# Initialize the YOLO model with a smaller, faster model
model = YOLO("yolov8n.pt")

# Object classes
classNames = model.names

# Function to get the direct stream URL from YouTube
def get_youtube_stream_url(youtube_url):
    video = pafy.new(youtube_url)
    best = video.getbest(preftype="mp4")
    return best.url

# Get the direct stream URL for YouTube video
#stream_url = 0  # Webcam as source
stream_url = get_youtube_stream_url('https://www.youtube.com/watch?v=KSsfLxP-A9g') #live stream
# Queue to hold frames
frame_queue = queue.Queue(maxsize=10)

# Function to capture frames from the camera
def capture_frames(camera, frame_queue):
    while True:
        success, frame = camera.read()
        frame = cv2.resize(frame, (480, 480)) 
        if not success:
            print("Failed to grab frame")
            break
        if frame_queue.full():
            frame_queue.get()
        frame_queue.put(frame)

# Function to process frames and run YOLO detection
def process_frames(frame_queue):
    while True:
        if not frame_queue.empty():
            frame = frame_queue.get()
            results = model(frame, show=False)
            
            for r in results:
                boxes = r.boxes
                for box in boxes:
                    x1, y1, x2, y2 = box.xyxy[0]
                    x1, y1, x2, y2 = int(x1), int(y1), int(x2), int(y2)
                    cv2.rectangle(frame, (x1, y1), (x2, y2), (255, 0, 255), 3)
                    confidence = math.ceil((box.conf[0] * 100)) / 100
                    cls = int(box.cls[0])
                    label = f'{classNames[cls]}: {confidence:.2f}'
                    org = (x1, y1 - 10 if y1 - 10 > 10 else y1 + 10)
                    font = cv2.FONT_HERSHEY_SIMPLEX
                    fontScale = 0.5
                    color = (255, 0, 0)
                    thickness = 1
                    cv2.putText(frame, label, org, font, fontScale, color, thickness, cv2.LINE_AA)

            cv2.imshow('YOLOv8 Live Detection', frame)

            if cv2.waitKey(1) & 0xFF == ord('q'):
                break

# Initialize video capture
camera = cv2.VideoCapture(stream_url)

# Start threads for capturing and processing frames
capture_thread = threading.Thread(target=capture_frames, args=(camera, frame_queue))
process_thread = threading.Thread(target=process_frames, args=(frame_queue,))

capture_thread.start()
process_thread.start()

capture_thread.join()
process_thread.join()

# Clean up
camera.release()
cv2.destroyAllWindows()
```
